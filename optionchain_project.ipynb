```python
#datapreprocessing
import pandas as pd
df = pd.read_csv('completedata.csv')

# Drop rows where 'name', 'expiry', or 'option_type' columns have missing values
df_cleaned = df.dropna(subset=['name', 'expiry', 'option_type'])

# Displaying the cleaned data
print(df_cleaned.head())

# save the cleaned data to a new CSV file
df_cleaned.to_csv('completepreprocesseddata.csv', index=False)
```

```python
df_cleaned.shape
```

```python
def get_option_chain_data(instrument_name: str, expiry_date: str, side: str) -> pd.DataFrame:
    # Convert expiry_date to date format
    expiry_date = pd.to_datetime(expiry_date, format='%d-%m-%Y').date()
    
    # Filtering using 'name'  and adjust the expiry date format
    options = df_cleaned[(df_cleaned['name'] == instrument_name) & 
                         (pd.to_datetime(df_cleaned['expiry'], format='%d-%m-%Y').dt.date == expiry_date) & 
                         (df_cleaned['option_type'] == side)]
    
    if options.empty:
        print(f"No matching options found for {instrument_name} with expiry {expiry_date} and side {side}.")
        return pd.DataFrame(columns=['instrument_name', 'strike_price', 'side', 'bid/ask'])

    # Collecting the highest prices for each strike
    output_data = []
    if side == 'PE':
        highest_bid = options['last_price'].max()
        for strike in options['strike'].unique():
            output_data.append({
                'instrument_name': instrument_name,
                'strike_price': strike,
                'side': side,
                'bid/ask': highest_bid
            })
    elif side == 'CE':
        highest_ask = options['last_price'].max()
        for strike in options['strike'].unique():
            output_data.append({
                'instrument_name': instrument_name,
                'strike_price': strike,
                'side': side,
                'bid/ask': highest_ask
            })

    output_df = pd.DataFrame(output_data)
    return output_df

# Usage
instrument_name = "NATURALGAS"
expiry_date = "23-12-2024"
side = "PE"

# Call function
option_chain_df = get_option_chain_data(instrument_name, expiry_date, side)
print("Option Chain Data:")
print(option_chain_df.to_string(index=False))

# Usage
instrument_name = "NATURALGAS"
expiry_date = "23-12-2024"
side = "CE"

# Call function
option_chain_df = get_option_chain_data(instrument_name, expiry_date, side)
print("Option Chain Data:")
print(option_chain_df.to_string(index=False))

```

```python
import pandas as pd

# Mock function to simulate fetching margin requirement from an API
def get_margin_requirement(instrument_name: str, strike_price: float, side: str) -> float:
    # Simulated margin requirement logic (for demonstration, using a constant or simple calculation)
    return strike_price * 0.10

def calculate_margin_and_premium(data: pd.DataFrame, lot_size: int) -> pd.DataFrame:
    # Check if the DataFrame is empty
    if data.empty:
        print("No data available to calculate margin and premium.")
        return data

    # Calculate margin required and premium earned
    data['margin_required'] = data.apply(
        lambda row: get_margin_requirement(row['instrument_name'], row['strike_price'], row['side']), axis=1)
    data['premium_earned'] = data['bid/ask'] * lot_size

    return data

# Example usage
lot_size = 10  # Assuming a lot size of 10 for this example
option_chain_df_with_margin = calculate_margin_and_premium(option_chain_df, lot_size)


# Print the resulting DataFrame
print("Option Chain Data with Margin and Premium:")
print(option_chain_df_with_margin.to_string(index=False))
```
